```python
### Convert Date Format to datetime
### Add Year and Month fields
df1.transaction_date = pd.to_datetime(df1.transaction_date, dayfirst=True) 
df1['year'] = df1.transaction_date.dt.year
df1['month'] = df1.transaction_date.dt.month
df1.head()
```

### Full Year Revenue Comparison

```python
### Total revenue by year (excluding current ytd)
#Find years included in dataframe
full_years = df1.groupby('year')['month'].nunique()
# Filter for years with all 12 months
fy = sorted(full_years[full_years == 12].index.tolist())
#print(fy)
df1_fullyears = df1[df1['year'].isin(fy)]
#df1_fullyears.head(50)
rev_fullyears = df1_fullyears.groupby(['year']).agg(total_revenue=('price', 'sum'))
rev_fullyears.head()
```

### YTD Comparison for all years

```python
### Total revenue for first 6 months of each year

# Group the dataframe by year and find the months for each year
months_per_year = df1.groupby('year')['month'].apply(set)

# Find the intersection of all month sets
fm = set.intersection(*months_per_year)

# Convert to a sorted list
fm = sorted(fm)

ytd=df1[df1.month.isin(fm)]
rev_ytd = ytd.groupby(['year']).agg(total_revenue=('price', 'sum'))
rev_ytd.head()
```

### Top 5 commodities of each year by count

```python
# Group by year and commodity, sort by count of occurrences
bc = df1.groupby(['year', 'commodity']).agg(total_purchases=('basket_id', 'count')) \
    .sort_values(['year', 'total_purchases'], ascending=[True, False]).reset_index()

# Group by year and take the top 5 commodities for each year
top5_cy = bc.groupby('year').head(5).reset_index(drop=True)
top5_cy['rank'] = top5_cy.groupby('year')['total_purchases'].rank(method='first', ascending=False)
#top5_cy.head(60)

# Pivot
pivoted_top5_cy = top5_cy.pivot(index='rank', columns='year', values='commodity')
pivoted_top5_cy.head()
```

### Top 5 commodities of each month by count

```python
# Group by month and commodity, sort by count of occurrences

bc_month = df1.groupby(['month', 'commodity']).agg(total_purchases=('basket_id', 'count')) \
    .sort_values(['month', 'total_purchases'], ascending=[True, False]).reset_index()


# Group by month and take the top 5 commodities for each month
top5_cm = bc_month.groupby('month').head(5).reset_index(drop=True)
top5_cm['rank'] = top5_cm.groupby('month')['total_purchases'].rank(method='first', ascending=False)
#top5_cm.head(60)
# Pivot
pivoted_top5_cm = top5_cm.pivot(index='rank', columns='month', values='commodity')
pivoted_top5_cm.head(50)
```

### Highest revenue commodities of each year

```python
# Group by year and commodity, sort by sum of price
rev_yr = df1.groupby(['year', 'commodity']).agg(total_revenue=('price', 'sum')) \
    .sort_values(['year', 'total_revenue'], ascending=[True, False]).reset_index()

# Group by year and take the top 5 commodities for each year
top5_ry = rev_yr.groupby('year').head(5).reset_index(drop=True)
top5_ry['rank'] = top5_ry.groupby('year')['total_revenue'].rank(method='first', ascending=False)
#top5_ry.head(60)

# Pivot
pivoted_top5_ry = top5_ry.pivot(index='rank', columns='year', values='commodity')
pivoted_top5_ry.head()
```

```python
# Loop through each unique year in the data
# Function to format y-axis as currency
def currency_format(x, _):
    return f'${x:,.2f}'  # Formats with a dollar sign, commas, and two decimal places


for year in fy:
# Filter data for the specific year
    year_data = top5_avg_prices[top5_avg_prices['year'] == year]
    
# Create the bar chart
    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=year_data,
        x='month',
        y='avg_price',
        hue='commodity'
    )
    
# Customize the chart
    plt.title(f'Average Price of Top 5 Commodities in {year}', fontsize=16)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Average Price', fontsize=12)
    plt.gca().yaxis.set_major_formatter(FuncFormatter(currency_format))  # Apply currency format to y-axis
    plt.legend(title='Commodity', loc='upper left', bbox_to_anchor=(1, 1))
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

# Display the chart
#    plt.tight_layout()
#    plt.show()
    
# Save the chart as an image file
    filename = f'.\\Charts\\avgprice_monthly-top5_{year}.png'  # Create a unique filename for each year
    plt.tight_layout()
    plt.savefig(filename, dpi=300)  # Save with high resolution (300 dpi)
    plt.close()  # Close the current figure to free memory

print("Charts Saved")    
```

```python
### Focus commodity stats
# Using commodity = 'deli meats', 'deli specialties', 'deli supplies', 'cheese', 'seafood-frozen', 
# 'seafood-misc', 'seafood-fresh', 'seafood-shelf stable', 'beef', 'pork', 'chicken', 'chicken/poultry', 
# 'pork', 'meat - other'

# Filter df1 for rows including focus products
focus_purchases = df1[df1['commodity'].isin(['deli meats', 'deli specialties', 'deli supplies', 'cheese', 'seafood-frozen', \
     'seafood-misc', 'seafood-fresh', 'seafood-shelf stable'])]

#Replace deli terms
focus_purchases['commodity'] = focus_purchases['commodity'].replace(['deli meats', 'deli specialties', 'deli supplies'], \
    'deli products')

#Replace seafood terms
focus_purchases['commodity'] = focus_purchases['commodity'].replace(['seafood-frozen', 'seafood-misc', 'seafood-fresh', \
     'seafood-shelf stable'], 'seafood')

#focus_purchases.head(50)

focus_stats = focus_purchases.groupby(['commodity']).agg(total_revenue=('price', 'sum'), count_of_purchases=('basket_id', 'count'), \
    pref_brand=('brand', lambda x: x.mode()[0]), pref_loyalty=('loyalty', lambda x: x.mode()[0]), pref_hh=('household_type', lambda x: x.mode()[0]), \
    pref_age=('age_band', lambda x: x.mode()[0])).sort_values('total_revenue', ascending=False).reset_index()
focus_stats.head(10)
```

```python
### Pie charts for Household Type purchases

household_color_map = {
    'single male': brand_colors['sea green'],
    'single female': brand_colors['chartreuse'],
    '1 adult with kids': brand_colors['deep sky blue'],
    '2 adults with kids': brand_colors['orchid'],
    '2 adults with no kids': brand_colors['lt periwinkle'],
}

# Group data by commodity and household type, and count the number of purchases
hh_purchase_counts = focus_purchases.groupby(['commodity', 'household_type']).size().reset_index(name='purchase_count')

# Calculate total purchases per commodity
total_hh_counts = purchase_counts.groupby('commodity')['purchase_count'].transform('sum')

# Calculate proportions
hh_purchase_counts['proportion'] = hh_purchase_counts['purchase_count'] / total_hh_counts

# Get unique commodities
commodities = hh_purchase_counts['commodity'].unique()

# Create pie charts
fig, axes = plt.subplots(1, len(commodities), figsize=(15, 5))  # Create subplots for pie charts

for idx, commodity in enumerate(commodities):
    # Filter data for the current commodity
    data = hh_purchase_counts[hh_purchase_counts['commodity'] == commodity]

    # Map household types to colors
    colors = [household_color_map[household] for household in data['household_type']]
    
    
    # Create pie chart
    axes[idx].pie(
        data['proportion'],  # Values
        labels=data['household_type'],  # Labels
        autopct='%1.1f%%',  # Display percentages
        startangle=140,  # Starting angle for better visualization
        colors=colors
    )
    # Add a title
    axes[idx].set_title(f'{commodity.title()} Purchases')

# Adjust layout
plt.suptitle('Proportion of Purchases by Household Type for Each Product', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the plot
plt.show()
```

```python
### Average spend per basket for each of the focus products for the last 6 months

# This will give the data we need to know how many additional purchases 
# we need to increase monthly in order to reach our revenue goals

# Find total basket count, total monthly revenue, and average basket revenue for each focus product for each month
focus_month_stats = focus_purchases.groupby(['year', 'month', 'commodity']).agg(total_baskets=('basket_id', 'nunique'), \
    total_revenue=('price', 'sum')).reset_index()
focus_month_stats['avg_basket_price']=focus_month_stats['total_revenue'] / focus_month_stats['total_baskets']
#focus_month_stats.head(50)

#Find average basket price for each focus product for the last 6 months
focus_month_stats_last6 = focus_month_stats.tail(18).reset_index(drop=True)
#focus_month_stats_last6.head(20)
focus_avgbp_last6 = focus_month_stats_last6.groupby(['commodity']).agg(avg_rev=('avg_basket_price', 'mean')).reset_index()
focus_avgbp_last6.head()                                                                   
```

```python
### Required revenue increase

# Using non-compounded values
# Decision to take an average of the last six months monthly revenue to calculate necessary 
# increases to mitigate effects of seasonality and ensure more realistic targets

# Monthly revenue in last month of data
tot_monthly_rev = tot_monthly_rev.sort_values(['year', 'month'])

# Limit to last 6 months
last_6 = tot_monthly_rev.tail(6)

avg_mr = last_6['monthly_rev'].mean()
req_increase = avg_mr * .025
print(f"Average monthly revenue for the last 6 months: ${avg_mr:,.2f}")
print(f"Required monthly revenue increase to achieve goal: ${req_increase:,.2f}")
```

```python
### Projected monthly revenue for the next 12 months to achieve 2.5% annual growth
revenue = avg_mr
for month in range(1, months + 1):
    revenue += req_increase
    print(f"Month {month}: ${revenue:,.2f}")
```

#### Goal will be to increase monthly revenue by $155.44

```python
### How many additional cheese, deli and seafood purchases do we need per month to achieve our goal?
### This will give me the total if I am only increasing purchases for a single product

# Take our monthly revenue increase goal and divide by average monthly basket for our focus products
focus_avgbp_last6['new_purchases'] = req_increase / focus_avgbp_last6['avg_rev']
focus_avgbp_last6['no_new_purchases'] = focus_avgbp_last6['new_purchases'].apply(math.ceil)
focus_avgbp_last6.head()
```

```python
### I want to work out what proportion of new purchases should be allocated to each product
focus_avgbp_last6_prop = focus_avgbp_last6

# Calculate the proportion of the total revenue goal for each product
focus_avgbp_last6_prop['proportional_goal'] = (focus_avgbp_last6['avg_rev'] / focus_avgbp_last6['avg_rev'].sum()) * req_increase

# Calculate the number of new purchases needed for each product
focus_avgbp_last6_prop['new_purchases'] = focus_avgbp_last6['proportional_goal'] / focus_avgbp_last6_prop['avg_rev']
focus_avgbp_last6_prop['new_purchases'] = focus_avgbp_last6['new_purchases'].apply(math.ceil)  # Round up to ensure the goal is met
focus_avgbp_last6_prop.head()
```
